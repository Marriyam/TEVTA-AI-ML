```python
X=[3,4,5,6,7]
y=[2.5, 3.2, 3.8, 6.5, 11.5]
mymodel = np.poly1d(np.polyfit(X, y, 3))
print(mymodel(X))
print(mymodel)
myline = np.linspace(1, 8, 100)
plt.scatter(X, y)
plt.plot(myline, mymodel(myline))
plt.show()
mse = mean_squared_error(y, mymodel(X))
print(r2_score(y, mymodel(X)))
print(f"Mean Squared Error: {mse}\n")

```

##Max level of Degree can be determined through the formula (n-1) Non-Linear Regression but for small datasets

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score

# Define the dataset
x = np.array([1, 2, 3, 4, 5, 6, 7, 8])  # Input features
y = np.array([2, 3, 5, 4, 6, 7, 8, 9])  # Target values

# Linear fit (Degree 1)
polynomial1 = np.poly1d(np.polyfit(x, y, 1))
x_line = np.linspace(min(x), max(x), 100)
y_line1 = polynomial1(x_line)
mse = mean_squared_error(y, polynomial1(x))
print("Degree 1:")
print(f"R2 Score: {r2_score(y, polynomial1(x))}")
print(f"Mean Squared Error: {mse}\n")
plt.scatter(x, y, color='red', label='Data Points')
plt.plot(x_line, y_line1, label='Degree 1 Polynomial')
plt.title('Polynomial Regression (Degree 1)')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()

# Quadratic fit (Degree 2)
polynomial2 = np.poly1d(np.polyfit(x, y, 2))
y_line2 = polynomial2(x_line)
mse = mean_squared_error(y, polynomial2(x))
print("Degree 2:")
print(f"R2 Score: {r2_score(y, polynomial2(x))}")
print(f"Mean Squared Error: {mse}\n")
plt.scatter(x, y, color='red', label='Data Points')
plt.plot(x_line, y_line2, label='Degree 2 Polynomial')
plt.title('Polynomial Regression (Degree 2)')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()

# Cubic fit (Degree 3)
polynomial3 = np.poly1d(np.polyfit(x, y, 3))
y_line3 = polynomial3(x_line)
mse = mean_squared_error(y, polynomial3(x))
print("Degree 3:")
print(f"R2 Score: {r2_score(y, polynomial3(x))}")
print(f"Mean Squared Error: {mse}\n")
plt.scatter(x, y, color='red', label='Data Points')
plt.plot(x_line, y_line3, label='Degree 3 Polynomial')
plt.title('Polynomial Regression (Degree 3)')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()

# Degree 4 fit
polynomial4 = np.poly1d(np.polyfit(x, y, 4))
y_line4 = polynomial4(x_line)
mse = mean_squared_error(y, polynomial4(x))
print("Degree 4:")
print(f"R2 Score: {r2_score(y, polynomial4(x))}")
print(f"Mean Squared Error: {mse}\n")
plt.scatter(x, y, color='red', label='Data Points')
plt.plot(x_line, y_line4, label='Degree 4 Polynomial')
plt.title('Polynomial Regression (Degree 4)')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()

# Degree 5 fit
polynomial5 = np.poly1d(np.polyfit(x, y, 5))
y_line5 = polynomial5(x_line)
mse = mean_squared_error(y, polynomial5(x))
print("Degree 5:")
print(f"R2 Score: {r2_score(y, polynomial5(x))}")
print(f"Mean Squared Error: {mse}\n")
plt.scatter(x, y, color='red', label='Data Points')
plt.plot(x_line, y_line5, label='Degree 5 Polynomial')
plt.title('Polynomial Regression (Degree 5)')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()

# Degree 6 fit
polynomial6 = np.poly1d(np.polyfit(x, y, 6))
y_line6 = polynomial6(x_line)
mse = mean_squared_error(y, polynomial6(x))
print("Degree 6:")
print(f"R2 Score: {r2_score(y, polynomial6(x))}")
print(f"Mean Squared Error: {mse}\n")
plt.scatter(x, y, color='red', label='Data Points')
plt.plot(x_line, y_line6, label='Degree 6 Polynomial')
plt.title('Polynomial Regression (Degree 6)')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()

# Degree 7 fit
polynomial7 = np.poly1d(np.polyfit(x, y, 7))
y_line7 = polynomial7(x_line)
mse = mean_squared_error(y, polynomial7(x))
print("Degree 7:")
print(f"R2 Score: {r2_score(y, polynomial7(x))}")
print(f"Mean Squared Error: {mse}\n")
plt.scatter(x, y, color='red', label='Data Points')
plt.plot(x_line, y_line7, label='Degree 7 Polynomial')
plt.title('Polynomial Regression (Degree 7)')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()

# Degree 8 fit
polynomial8 = np.poly1d(np.polyfit(x, y, 8))
y_line8 = polynomial8(x_line)
mse = mean_squared_error(y, polynomial8(x))
print("Degree 8:")
print(f"R2 Score: {r2_score(y, polynomial8(x))}")
print(f"Mean Squared Error: {mse}\n")
plt.scatter(x, y, color='red', label='Data Points')
plt.plot(x_line, y_line8, label='Degree 8 Polynomial')
plt.title('Polynomial Regression (Degree 8)')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

##Multiple Non-Linear Regression

```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Sample data
X = np.array([[1, 2], [2, 3], [3, 4], [4, 5]])
y = np.array([5, 7, 9, 11])

# Transform data to include polynomial terms
poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X)

# Fit the regression model
model = LinearRegression()
model.fit(X_poly, y)

# Predict and evaluate
y_pred = model.predict(X_poly)
mse = mean_squared_error(y, y_pred)

print(f"Coefficients: {model.coef_}")
print(f"Mean Squared Error: {mse}\n")
print(f"R2 Score: {r2_score(y, y_pred)}")
```
