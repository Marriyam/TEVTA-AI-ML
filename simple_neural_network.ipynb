```python
import numpy as np
x1=1
x2=1
w1=w2=0.01
```

```python
def network(w1, w2):
  return (w1*x1)+(w2*x2)
def output(net):
  return 1/(1+np.exp(-net))

net=network(w1,w2)
out=output(net)
previous_out=0
threshold = 0.0001
target=0.5
while abs(out - target) > threshold:
  net=network(w1,w2)
  out=output(net)
  Error=((out-target)*(out*(1-out))*x1)
  Error2=((out-target)*(out*(1-out))*x2)
  w1=w1-0.1*Error
  w2=w2-0.1*Error2
  print(w1)
  print(w2)
  print("output=",out)
  print("\n")
  previous_out=out
```
