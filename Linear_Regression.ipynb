###Linear Regression

Statistical method used to model the relationship between a dependent variable (target) and one or more independent variables (predictors).

###Equation:
we began with equation of straight line: *y=mx+c*


###Linear Regression with Scipy

```python
# Importing Libraries
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python

X=[1,2,3,4,5,6,8,9]
y=[4,7,9,11,12,15,17,20]

slope, intercept, r, p, std_err = stats.linregress(X, y) #
print(slope,intercept)
def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, X))

plt.scatter(X, y)  #it just plots xy discrete values and will not make a line
plt.plot(X, mymodel) #will draw continuous line
plt.show()
```

#### Calculating Slope & Y-Intercept

```python
X=[1,2,3]
y=[4,7,9]
n= len(X)
sum_x=0;
sum_y=0;
sum_xy=0;
sum_xx=0;
for i in range(n):
    sum_x=sum_x+X[i]
    sum_y=sum_y+y[i]
    sum_xy=sum_xy+(X[i]*y[i])
    sum_xx=sum_xx+(X[i]*X[i])

mean_X=sum_x/n
mean_y=sum_y/n
mean_xy=sum_xy/n
mean_xx=sum_xx/n
print(mean_X)
print(mean_y)
print(mean_xy)
print(mean_xx)
m=(mean_xy-(mean_X*mean_y))/(mean_xx-(mean_X*mean_X));
c=mean_y-(m*mean_X);
print(m,c)
```

###Linear Regression with Sklearn

```python
X = np.array([[1, 3, 5, 8, 9], [4, 8, 10, 14, 18 ]])
y = np.array([[22, 23, 29, 31, 32], [38, 37, 43.5, 44, 45]])
X = X.reshape(-1, 1)
y = y.reshape(-1, 1)
regression = LinearRegression()
regression.fit(X, y)
print(regression.score(X, y))
```

```python
y_pred = regression.predict(X)
plt.scatter(X, y_pred, color ='b')
plt.plot(X, y_pred, color ='k')

plt.show()
```
