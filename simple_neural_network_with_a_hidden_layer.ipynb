```python
x1 = 1
x2 = 1

w1 = w3 = w6 = 0.5
w2 = w4 = w5 = 1

threshold = 0.0001
target = 0
X = 0.01

output = 1
while abs(output - target) > threshold:
    h1out = (w1 * x1) + (w2 * x2)
    h2out = (w3 * x1) + (w4 * x2)
    output = (w5 * h1out) + (w6 * h2out)

    error = output - target

    w5 -= X * error * h1out
    w6 -= X * error * h2out
    w1 -= X * error * w5 * x1
    w2 -= X * error * w5 * x2
    w3 -= X * error * w6 * x1
    w4 -= X * error * w6 * x2

    print("Output:", output)

print("\n\n Output:", output)
print("Weights:", w1, w2, w3, w4, w5, w6)
```
